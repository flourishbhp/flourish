```python
import os
import sys

import django

sys.path.append('../')  # add path to project root dir

os.environ["DJANGO_SETTINGS_MODULE"] = "flourish.settings"
from django.apps import apps as django_apps

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

django.setup()
```

```python
from flourish_caregiver.models import OnScheduleCohortATb6Months
from edc_appointment.models import Appointment
from edc_visit_schedule.models import SubjectScheduleHistory

caregiver_subject_identifier = 'B142-040990128-6'
child_subject_identifier = f'{caregiver_subject_identifier}-60'
schedule_name = 'a_tb1_6_months_schedule1'


try:
    onschedule_obj = OnScheduleCohortATb6Months.objects.get(
        subject_identifier = caregiver_subject_identifier,
        child_subject_identifier = child_subject_identifier,
        schedule_name = 'a_tb1_6_months_schedule1'
    )
except OnScheduleCohortATb6Months.DoesNotExist:
    pass
else:
    onschedule_obj.schedule_name = 'a_tb2_6_months_schedule1'
    onschedule_obj.save()
    
    Appointment.objects.filter(
        subject_identifier = caregiver_subject_identifier,
        schedule_name=schedule_name
    ).update(schedule_name='a_tb2_6_months_schedule1')
    
    SubjectScheduleHistory.objects.filter(
        subject_identifier = caregiver_subject_identifier,
        schedule_name = schedule_name, 
    ).update(schedule_name='a_tb2_6_months_schedule1')
    


```
