```python
import os, sys
import django
import datetime
from django.apps import apps as django_apps
sys.path.append('../') # add path to project root dir
os.environ["DJANGO_SETTINGS_MODULE"] = "flourish.settings"
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
```

```python
from flourish_child.models import Appointment as ChildAppointment
from edc_appointment.models import Appointment

lower_dt = datetime.date(2024, 1, 1)
schedules = SubjectScheduleHistory.objects.filter(schedule_name__icontains='sq_fu',
                                                  onschedule_model__startswith='flourish_child',
                                                  onschedule_datetime__date__lt=lower_dt)
count = 0
for schedule in schedules:
    subject_identifier = schedule.subject_identifier
 
    child_visits = ChildVisit.objects.filter(subject_identifier=subject_identifier,
                                             schedule_name=schedule.schedule_name)
    if not child_visits.exists():
        print(subject_identifier)
        child_consent = CaregiverChildConsent.objects.filter(
            subject_identifier=subject_identifier).earliest('consent_datetime')
        caregiver_sidx = child_consent.subject_consent.subject_identifier
        cohort = Cohort.objects.filter(
            subject_identifier=subject_identifier, current_cohort=True).latest('assign_datetime')
        c_schedule = CohortSchedules.objects.get(schedule_type='sq_followup',
                                                 cohort_name=cohort.name,
                                                 child_count=child_consent.caregiver_visit_count)
        maternal_visits = MaternalVisit.objects.filter(subject_identifier=caregiver_sidx,
                                                       schedule_name=c_schedule.schedule_name)
        if not maternal_visits.exists():
            child_onschedule_cls = django_apps.get_model(schedule.onschedule_model)
            caregiver_onschedule_cls = django_apps.get_model(c_schedule.onschedule_model)
            child_onschedules = child_onschedule_cls.objects.filter(
                subject_identifier=subject_identifier, schedule_name=schedule.schedule_name)
            caregiver_onschedules = caregiver_onschedule_cls.objects.filter(
                subject_identifier=caregiver_sidx, schedule_name=c_schedule.schedule_name)
            caregiver_ssh = SubjectScheduleHistory.objects.filter(
                subject_identifier=caregiver_sidx, schedule_name__in=caregiver_onschedules.values_list(
                    'schedule_name', flat=True))
            child_appts = ChildAppointment.objects.filter(
                subject_identifier=subject_identifier, schedule_name__in=child_onschedules.values_list(
                    'schedule_name', flat=True))
            caregiver_appts = Appointment.objects.filter(
                subject_identifier=caregiver_sidx, schedule_name__in=caregiver_onschedules.values_list(
                    'schedule_name', flat=True))
            caregiver_appts.delete()
            child_appts.delete()
            caregiver_ssh.delete()
            caregiver_onschedules.delete()
            schedule.delete()
            child_onschedules.delete()
```
