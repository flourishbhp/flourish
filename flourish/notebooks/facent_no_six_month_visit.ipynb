Facet participnats with no six month Visit

```python
from flourish_facet.models import FacetConsent,FacetVisit
import csv

facet_visits = FacetVisit.objects.values_list('subject_identifier',flat=True)
facet_consents_no_visits= FacetConsent.objects.exclude(subject_identifier__in=facet_visits).values_list('subject_identifier',flat=True)

with open ('participants_no_visit.csv','w',newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['subject_identifiers'])
    for i in range((facet_consents_no_visits.count())):
        writer.writerow([facet_consents_no_visits[i]])
```

Facet eligible participants with no screening

```python
from flourish_caregiver.models import SubjectConsent
from flourish_facet.views import EligibleFacetParticipantsMixin
from flourish_facet.models import FacetSubjectScreening

flourish_consents = SubjectConsent.objects.all()

facet_eligible = EligibleFacetParticipantsMixin()
facet_eligible_participants =facet_eligible.eligible_participants(flourish_consents)

facet_screened = FacetSubjectScreening.objects.values_list('subject_identifier',flat=True)

unscreened_participants = facet_eligible_participants.exclude(subject_identifier__in=facet_screened).values_list('subject_identifier',flat=True)

with open ('eligible_participants.csv','w',newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['subject_identifiers'])
    for i in range((unscreened_participants.count())):
        writer.writerow([unscreened_participants[i]])
```
